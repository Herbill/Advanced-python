```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel("data.xlsx")
df.head()
```

```python
df["Date"] = pd.to_datetime(df["Date"])  

plt.plot(df["Date"], df["Sales"], color = 'green' )
plt.title('Sales over time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.bar(df["Product"], df['Sales'])
plt.title('Product Sales over time')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()
```

```python
df=pd.read_excel("data.xlsx")
df
```

```python
sales_data =  df.groupby('Category')['Sales'].sum()
sales_data
```

```python
# to create a pie chart

plt.figure(figsize=(8, 8))
plt.pie(sales_data, labels=sales_data.index, autopct='%1.1f%%')

plt.title('proportion of sale by property')
plt.show()
```

```python
plt.hist(df['Sales'], bins=12, color='blue', edgecolor='black')

plt.title('Histogram')
plt.xlabel('Sales')
plt.ylabel('Frequencies')
plt.show()
```

```python
plt.fill_between(df['Date'], df['Sales'], color='lightblue', alpha=0.7)
plt.plot(df["Date"], df["Sales"], color = 'blue' )
plt.title('Sales over time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```
